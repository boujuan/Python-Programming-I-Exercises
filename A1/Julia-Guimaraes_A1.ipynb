```python
# importing only the function, which creates random num
from numpy.random import randint
#asking the user for an input 
qs = int(input("Give me an number btw. 0 and 100"))
ll = 0 #starting value for python to guess
hl = 100 #ending guess num:higher limit
a = 0 #starting point for iteration
rn = randint(ll,hl) #random number, which python guesses 
while rn != qs: # as long rn is not qs 
    a += 1 #equals a = a + 1; do another iteration
    if rn > qs: 
        #if py guess is larger than our input 
        #set the guess (tn) as new highest value (hl)
        # delete everything btw 100 and rn 
        print(rn,"is larger than",qs)
        hl = rn 
        # update the numerical quantity 
        # with the new hl
        # rn = ran. gives us a new guess 
        #with the new values for hl 
        rn = randint(ll, hl) 
    elif rn < qs:
        #if rn is smaller than our input 
        #set the lowest input = ll = rn 
        #forget everything btw 0 and rn 
        print(rn, "is smaller than",qs)
        ll = rn
        rn = randint(ll, hl) 
print(rn,"equals",qs)
print("I needed", a, "guesses for that.")
```

```python
from numpy.random import randint
# define a class for the vector structure 
class VecClass: 

    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        self.magnitude = self.magnitude(x, y, z)
    
    def magnitude(self, x,y,z):
        rmg = (x**2 + y**2 + z**2)**(1/2)
        #rmg = round(mg, 2)
        return rmg 
#some notes about return statement below!

#create the main class - where the list of 10 
# vectors is created

n = 0 #counter for the vec + mg, starting with 0 
vecl = [] # This is the vector list (empty)
for i in range(10):
    #creating vector using vecclass  till we repeated that 1o times
    ovec = VecClass(randint(0,30), randint(0,30), randint(0,30))
    # appending this 10 vec to the empty list 
    vecl.append(ovec) #ovec = one vector 
    n += 1 #counting trough all vec and mag 
    # printing all mg and vec
    print("This is vector",n, ":", ovec.x, ovec.y, ovec.z)
    print(f"This is magnitude {n}: {ovec.magnitude:.2f}")
    #.2f is printing only the last 2 decimals 
    # for scientific notation its: .2e
    
# i is an object of the class = VecClass 
# for each i = vec it is printing the parameters
# we are telling the interpreter what the object  
# inside of the class means 

nm = 0 #counter for magnitudes
allmg = [] #empty list for magnitudes
for i in vecl: #for each vec in the vec list
    allmg.append(i.magnitude) #append the magnitude to 
    #a new list allmg 
    nm += 1 #numeration of each mg 
#print(allmg)

#find max and min magnitudes 
max_mag = max(allmg)
min_mag = min(allmg)
#numerate max and min magnitudes = giving them index
max_mag_i = allmg.index(max_mag)
min_mag_i = allmg.index(min_mag)

#define the max vector with the given index 
vecx = vecl[max_mag_i].x 
vecy = vecl[max_mag_i].y
vecz = vecl[max_mag_i].z

#define the min vector with the given index 
vicx = vecl[min_mag_i].x 
vicy = vecl[min_mag_i].y
vicz = vecl[min_mag_i].z

#time to print them all 
#only works because max_vec 
#has the same index than max_mg 
print(f"This is max vec: {vecx, vecy, vecz}")
print(f"This the max mg of vec: {max_mag}")
print(f"This is min vec: {vicx, vicy, vicz}")
print(f"This the max mg of vec: {min_mag}")
```

```python
def sum(x,n): #defining the sum function 
    sumx = 0 
    for k in range(n): 
       f =  x**k
       sumx = sumx + f 
    return sumx
#for the sum function we need to iterate for each N ones 
#in each iteration one sum of the actual N is added to..
#..the sum of N-1 (sumx = sumx +f)

def geo(x,n): #defining the geometric series
    return  (1-x**n)/(1-x)

#xl = [0.1, 0.2, 0.3, 0.4]
xl = [0.1, 0.2] #values for x
N = 50 #end threshold for su or geo 
k = 0 #starting+step point for the calculation 
l = []
while k < N:
    k += 1
    for x in xl: 
        if geo(x,k) != sum(x,k):
            print(f"Geometric: {geo(x,k)} Sum: {sum(x,k)}")
            #print(f"sum - geometric:{sum(x,k)-geo(x,k)}")
            l.append(geo(x,k))
            l.append(sum(x,k))

print(f"If geo would be always unequal with sum, the list l")
print(f"would have (len(xl) = 2)*(N=50)*(sum+geo=2)=200 floats stored") 
print(f"but the length of l is: {len(l)}")
print(f"So {200 - len(l)} results are equal")        
# The results of the geometric series are more precise
# than the sum due to rounding errors in 
# every iteration 
```

# Exercise 5
Trapezoidal rule:
$$ \int_{a}^{b}f(x)dx â‰ˆ h \left( \frac{f\left( x_n \right) + f\left( x_0 \right) }{2}+\sum_{k=1}^{N-1}f\left( a+h\cdot k\right) \right) $$ 

with $$ h = \frac{b-a}{n}$$



```python
import math

# define the function f(x) = sin(x) where [x]= degrees
def f(x):
    # Convert degrees to radians using math.radians()
    x = math.radians(x)
    return math.sin(x)

# Define the function g(x) = cos(x) 
def g(x):
    # Convert degrees to radians using math.radians()
    x = math.radians(x)
    return math.cos(x)

il = [] # list with all intersections 
for i in range(0, 361):
    if (abs(f(i) - g(i)) < 0.00001):
        il.append(i)
        
print("Angles where f(x) and g(x) intersect:", il)

def j(x):
    return x**2+2*x-5

def k(x):
    return 2*x-1

il2 = [] #the list where we store the intersections
for i in range(-5, 6):
   if j(i) == k(i):
       il2.append(i) #append all values to il2,
       #where j and k intersect 

print("The intersections between j(x) and k(x) are:" ,il2)
 


#define the trapezoidal rule:
def tr(f,a, b, n): #starting = a end = b -point [-2, 2] here
    h = (b-a)/n # h is the width of every integral part  
    integral = 0.5* (f(a) + f(b)) #integral formula...
    #...without the sum, following now:
    for i in range(1, n):
        sum = f(a + i*h)
        integral = integral + sum
    return integral *h #multiplying the full thing with h 

n = 100  #setting the number of integral parts 
resultj = abs(tr(j, il2[0], il2[1],n))
resultk = abs(tr(k, il2[0], il2[1],n))
print(f"The area of j(x) between a und b is: {resultj:.2f}")
print(f"The area of k(x) between a und b is: {resultk:.2f}")
print(f"The area between j(x) and k(x) inside of intersections")
print(f"a and b is {resultj - resultk:.2f}")

```

```python
# Generate cloud of points
def generate_points():
    import numpy as np
    x = np.linspace(0, 10, 1000)
    noise = np.random.normal(0, 3, len(x))
    a,b = np.random.randint(1, 5), np.random.randint(-3, 3)
    y = a * x + b + noise
    print(f"Original function: y = {a}x + ({b})")
    return x, y

#defining the coefficient b 
def b_(x,y):
    n = len(x)
    return (sum(y)/n) - a_(x,y) * (sum(x)/n)

#defining the coefficient a 
def a_(x,y):
    meanx = (sum(x))/len(x)
    meany = (sum(y))/len(y)
    num = 0 
    den = 0
    for i in range(len(x)):
       num +=  (x[i] - meanx)*(y[i] - meany)
       den += (x[i] - meanx)**2
    return num / den

#generate the cloud with x a and b 
x, y = generate_points() 
#calling / executing function b 
coefb = b_(x,y)
#calling / executing function a 
coefa = a_(x,y)

print(f"The guessed function: y = {coefa:.2f}x  +({coefb:.2f})")
```

```python
#defining the function 
#..which calculates the tree 
def tree(height):
    for i in range(height-1):
        spaces = (" "*abs(i-height+2))
        leaves = ("*"*(((i)*2)+1))
        print(f"{spaces}{leaves}")

#define the trunk width 
    trunkw = int(1/3 * height)

#define the position of the trunk 
    for _ in range(trunkw):
        spaces = (" " * int((i - trunkw//2))) #def
        trunk = ("|" * trunkw) #def
        print(f"{spaces}{trunk}") #print 

height = input("give me a positive number:")
#height= 10 #test 
tree(int(height)) #execute all 

```
